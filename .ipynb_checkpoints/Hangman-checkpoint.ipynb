```python
wordBank = ["fish", "platypus", "xylophone", "flip-flop", "Donkey Kong"]
```

```python
#Choose a random word
import random

def wordChoice():
    return wordBank[random.randint(0,len(wordBank)-1)]

word = wordChoice()
```

```python
#Turn word choice into line of _
def makeBlank():
    listWord = list(word)
    blankWord = listWord
    for i in range(0,len(listWord)):
        if blankWord[i] == " ":
            blankWord[i] = " "
        elif blankWord[i] == "-":
            blankWord[i] = "-"
        else:
            blankWord[i] = "_"
    return ' '.join(blankWord)

makeBlank()
```

```python
#Guess the word
def guessWord():
    hangman = 0
    while hangman <= 8:
        correct = False
        guess = input("Please input your guess").lower()
        for i in range(0,len(listWord)):
            if listWord[i] == guess:
                blank[i] = guess
                correct = True
        print(' '.join(blank))
        if correct == True:
            print("Correct!")
            for i in range(0,len(listWord)):
                if blank[i] != "_":
                    "You win!"
                    break
        else:
            print("Incorrect!")
            hangman = hangman + 1
            if hangman == 1:
                print("\n","\n","\n","\n","\n","  _________")
            elif hangman == 2:
                print("\n",
                      "||\n",
                      "||\n",
                      "||\n",
                      "||\n",
                      "||_________")
            elif hangman == 3:
                print("__________\n",
                      "||\n",
                      "||\n",
                      "||\n",
                      "||\n",
                      "||_________")
            elif hangman == 4:
                print("__________\n",
                      "||    ¦\n",
                      "||\n",
                      "||\n",
                      "||\n",
                      "||_________")
            elif hangman == 5:
                print("__________\n",
                      "||    ¦\n",
                      "||    O\n",
                      "||\n",
                      "||\n",
                      "||_________")
            elif hangman == 6:
                print("__________\n",
                      "||    ¦\n",
                      "||    O\n",
                      "||    | \n",
                      "||\n",
                      "||_________")
            elif hangman == 7:
                print("__________\n",
                      "||    ¦\n",
                      "||    O\n",
                      "||   /|\n",
                      "||\n",
                      "||_________")
            elif hangman == 8:
                print("__________\n",
                      "||    ¦\n",
                      "||    O\n",
                      "||   /|\ \n",
                      "||   /\n",
                      "||_________")
            else:
                print("__________\n",
                      "||    ¦\n",
                      "||    O\n",
                      "||   /|\ \n",
                      "||   / \ \n",
                      "||_________")
                print("Game Over")
                return True
```

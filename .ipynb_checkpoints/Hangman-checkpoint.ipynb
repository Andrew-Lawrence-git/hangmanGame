```python
#Choose a random word
import random

def wordChoice():
    '''Returns a random word from the wordbank'''
    wordBank = ["fish", "platypus", "xylophone", "flip-flop", "Donkey Kong"]
    return wordBank[random.randint(0,len(wordBank)-1)]
```

```python
#Turn word choice into underscores of correct length
def makeBlank():
    '''Converts chosen word into a list of underscores'''
    listWord = list(word)
    blankWord = listWord
    for i in range(0,len(listWord)):
        if blankWord[i] == " ":
            blankWord[i] = " "
        elif blankWord[i] == "-":
            blankWord[i] = "-"
        else:
            blankWord[i] = "_"
    return blankWord
```

```python
#Guess the word
def guessWord():
    '''Provides input for a letter, checks that letter against the chosen word, return either an extra part of the hangman, updating the letter or providing a win.'''
    global hangman
    global win
    hangman = 0
    win = 0
    while hangman <= 8:
        correct = False
        guess = input("Please input a letter").lower() #Player inputs a letter
        if guess.isalpha() == False or len(guess) > 1: #If the letter wasn't alphabetical or was longer than 1 character return an error
            print("Error: either your guess was over one character or it was not a letter. Please try again.")
        else:
            for i in range(0,len(listWord)): #Check the guess against each letter in the word
                if listWord[i] == guess:
                    blank[i] = guess
                    correct = True
            print(' '.join(blank)) #Display word with correct guesses filled in
            if win >= len(listWord)-1: #If the number of correct letters is equal to or greater than the number of letters in the word, win the game
                print("You win!")
                break   
            if correct == True: #If the letter was correct, increase win counter for each correct letter
                print("Correct!")
                win = 0
                for i in range(0,len(listWord)):
                    if blank[i] != "_":
                        win = win + 1
            else: #If the letter wasn't correct, increase hangman counter and print the current hangman
                print("Incorrect!")
                hangman = hangman + 1
                if hangman == 1:
                    print("\n","\n","\n","\n","\n","  _________")
                elif hangman == 2:
                    print("\n",
                          "||\n",
                          "||\n",
                          "||\n",
                          "||\n",
                          "||_________")
                elif hangman == 3:
                    print("__________\n",
                          "||\n",
                          "||\n",
                          "||\n",
                          "||\n",
                          "||_________")
                elif hangman == 4:
                    print("__________\n",
                          "||    ¦\n",
                          "||\n",
                          "||\n",
                          "||\n",
                          "||_________")
                elif hangman == 5:
                    print("__________\n",
                          "||    ¦\n",
                          "||    O\n",
                          "||\n",
                          "||\n",
                          "||_________")
                elif hangman == 6:
                    print("__________\n",
                          "||    ¦\n",
                          "||    O\n",
                          "||    | \n",
                          "||\n",
                          "||_________")
                elif hangman == 7:
                    print("__________\n",
                          "||    ¦\n",
                          "||    O\n",
                          "||   /|\n",
                          "||\n",
                          "||_________")
                elif hangman == 8:
                    print("__________\n",
                          "||    ¦\n",
                          "||    O\n",
                          "||   /|\ \n",
                          "||   /\n",
                          "||_________")
                else:
                    print("__________\n",
                          "||    ¦\n",
                          "||    O\n",
                          "||   /|\ \n",
                          "||   / \ \n",
                          "||_________")
                    print("Game Over")
```

```python
wordBank = ["fish", "platypus", "xylophone", "flip-flop", "Donkey Kong"]
```

```python
#Choose a random word
import random

def wordChoice():
    return wordBank[random.randint(0,len(wordBank)-1)]

word = wordChoice()
```

```python
#Turn word choice into line of _
def makeBlank():
    listWord = list(word)
    blankWord = listWord
    for i in range(0,len(listWord)):
        if blankWord[i] == " ":
            blankWord[i] = " "
        elif blankWord[i] == "-":
            blankWord[i] = "-"
        else:
            blankWord[i] = "_"
    return ' '.join(blankWord)

makeBlank()
```
